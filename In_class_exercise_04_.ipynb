##**1.1 Basic feature extraction using text data**

```python
from six.moves import urllib
import pandas as pd

text_array = []
url = "https://raw.githubusercontent.com/unt-iialab/INFO5731_FALL2020/master/In_class_exercise/01-05-1%20%20Adams%20v%20Tanner.txt"
text_file = urllib.request.urlopen(url)

for line in text_file:
  decoded_data = line.decode("utf-8").replace('\r\n', '')
  if decoded_data: text_array.append(decoded_data)
df = pd.DataFrame (text_array, columns = ['Raw Text'])
print(df)
```

**Number of sentences**

```python
df['Number of Sentences'] = df['Raw Text'].apply(lambda x: len(str(x).split(".")))
df
```

**Number of words**

```python
df['Number of Words'] = df['Raw Text'].apply(lambda x: len(str(x).split(" ")))
df
```

**Number of characters**

```python
df['Number of Characters'] = df['Raw Text'].str.len()
df
```

**Average word length**

```python
def avg_word_length(sentence):
  words = sentence.split()
  if len(words) != 0:
    return(sum(len(word) for word in words)/len(words))
  else:
    return None

df['Average Word Length'] = df['Raw Text'].apply(lambda x: avg_word_length(x))
df
```

**Number of stopwords**

```python
from nltk.corpus import stopwords

import nltk
nltk.download('stopwords')
```

```python
stop = stopwords.words('english')

df['Number of Stopwords'] = df['Raw Text'].apply(lambda x: len([x for x in x.split() if x in stop]))
df
```

**Number of special characters**

```python
def special_characters(sentence):
  special = 0
  for value in range(len(sentence)):
    if not((sentence[value].isalpha()) and (sentence[value].isdigit())):
        special = special + 1
  return special

df['Number of special characters'] = df['Raw Text'].apply(lambda x: special_characters(x))
df
```

**Number of numerics**

```python
df['Numbers of numerics'] = df['Raw Text'].apply(lambda x: len([x for x in x.split() if x.isdigit()]))
df
```

**Number of uppercase words**

```python
df['Number of upper case words'] = df['Raw Text'].apply(lambda x: len([x for x in x.split() if x.isupper()]))
df
```

## **1.2 Basic Text Pre-processing of text data**

**Lower casing**

```python
df['Lower Case'] = df['Raw Text'].apply(lambda x: " ".join(x.lower() for x in x.split()))
df
```

**Punctuation removal**

```python
df['Removal of Punctuation'] = df['Lower Case'].str.replace('[^\w\s]','')
df
```

**Stopwords removal**

```python
from nltk.corpus import stopwords
stop = stopwords.words('english')
df['Stopwords Removal'] = df['Removal of Punctuation'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
df
```

**Frequent words removal**

```python
freq_words = pd.Series(' '.join(df['Stopwords Removal']).split()).value_counts()[:10]
print(freq_words)
freq_words_list = list(freq_words.index)
df['After common words removal'] = df['Stopwords Removal'].apply(lambda x: " ".join(x for x in x.split() if x not in freq_words_list))
df
```

**Rare words removal**

```python
rare_words = pd.Series(' '.join(df['Stopwords Removal']).split()).value_counts()[-10:]
print(rare_words)
rare_words_list = list(rare_words.index)
df['After Rare Words removal'] = df['After common words removal'].apply(lambda x: " ".join(x for x in x.split() if x not in rare_words_list))
df
```

**Spelling correction**

```python
from textblob import TextBlob
df['Spelling Correction'] = df['After Rare Words removal'].apply(lambda x: str(TextBlob(x).correct()))
df
```

**Tokenization**

```python
import nltk
nltk.download('punkt')
df['Tokenization'] = df['Spelling Correction'].apply(lambda x: TextBlob(x).words)
df
```

**Stemming**

```python
from nltk.stem import PorterStemmer
st = PorterStemmer()
df['After Stemming'] = df['Tokenization'].apply(lambda x: " ".join([st.stem(word) for word in x]))
df
```

**Lemmatization**

```python
from textblob import Word
import nltk
nltk.download('wordnet')

df['After Lemmatization'] = df['After Stemming'].apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()]))
df
```

## **1.3 Saving to CSV**

```python
df.to_csv('cleaned_data.csv',index=False)
```

## **1.4 Advance Text Processing**

**Calculating Term frequencies**

```python
tf1 = (df['After Lemmatization']).apply(lambda x: pd.value_counts(x.split(" "))).sum(axis = 0).reset_index()
tf1.columns = ['words','tf']
tf1
```

**Top 10 - one grams**

```python
import itertools
import collections
from nltk import ngrams
from nltk.tokenize import word_tokenize

words_sentence = []
for sentence in df['After Lemmatization']:
  words_sentence.append(word_tokenize(sentence))
aftercleaning = [x for x in words_sentence if x != []]
iterations = list(itertools.chain.from_iterable(aftercleaning))
one_grams = ngrams(iterations, 1)
collections.Counter(one_grams).most_common(10)
```

**Top 10 - Bi grams**

```python
bi_grams = nltk.bigrams(iterations)
frequency_dist = nltk.FreqDist(bi_grams)
(frequency_dist.most_common(10))
```

**Top 10 - Tri grams**

```python
tri_grams = nltk.trigrams(iterations)
frequency_dist = nltk.FreqDist(tri_grams)
frequency_dist.most_common(10)
```
